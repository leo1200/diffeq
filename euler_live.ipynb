# Simple Example of the Explicit Euler Method

```julia
using Plots

# Setting starting conditions
y0 = 1
alpha = 1

# Specify parameters
dt = 2.5

# time
t = 0:dt:10
y = zeros(length(t))
y[1] = y0

# Euler's method
for i in 1:length(t)-1
    y[i+1] = y[i] - alpha * dt * y[i]
end

# Plotting
plot(t, y, label="Euler's method", lw = 3)

tfine = 0:0.01:10
plot!(tfine, exp.(-alpha * tfine), label = "Exact solution", lw = 3)
xlabel!("t")
ylabel!("y(t)")
title!("Solution to dy/dt = -y \n using Forward Euler Method with dt = $dt")
```
